```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'Admission.csv') #importing the data set
```

```python
df.head()
```

```python
df.dtypes #showing the datatype of each column
```

```python
df.shape#shape of dataset
```

```python
#updating chance of admission value
df.loc[df['Chance of Admit ']<0.8,'Chance of Admit ']=0
df.loc[df['Chance of Admit ']>=0.8,'Chance of Admit ']=1
```

```python
df
```

```python
#displaying max and min scores
print("Max GRE Score= ",df['GRE Score'].max())
print("Min GRE Score= ",df['GRE Score'].min())
```

```python
df=df.drop(['Serial No.'],axis=1)#dropping the serial no. column
```

```python
df
```

```python
x=df.drop('Chance of Admit ',axis=1)#taking all columns except one as X
y=df['Chance of Admit ']# taking chnce of admit as y
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
print("Number of students with higher chance of admission= ",df.loc[df['Chance of Admit ']==1,'Chance of Admit '].count())
```

```python
print("Number of students with Resaech papers= ",df.loc[df['Research']==1,'Research'].count())
```

```python
print("Number of students with good TOEFL score= ",df.loc[df['TOEFL Score']>=110,'TOEFL Score'].count())
```

```python
from sklearn.tree import DecisionTreeClassifier #importing decision tree 
model=DecisionTreeClassifier()
model=model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test) #predicting the value
```

```python
y_pred
```

```python
y_test
```

```python
#calculating various scores
from sklearn import metrics
print("Accuracy score= ",metrics.accuracy_score(y_test,y_pred))
print("Precision score= ",metrics.precision_score(y_test,y_pred))
print("Recall score= ",metrics.recall_score(y_test,y_pred))
print("F1 score= ",metrics.f1_score(y_test,y_pred))
```

```python
#calculating confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO 
from IPython.display import Image  
import pydotplus
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = ['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research'],class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Admission.png')
Image(graph.create_png())
```
